```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('/Users/gary/opt/anaconda3/images/266_sld_w.png')
plt.imshow(im)
```

```python
gray_im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_im, cmap="gray", vmin=0, vmax=255)
```

```python
gray_correct = np.array(255 * (gray_im / 255) ** 1.2 , dtype='uint8')
plt.imshow(gray_correct, cmap="gray", vmin=0, vmax=255)
```

```python
thresh = cv2.adaptiveThreshold(gray_correct, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 255, 1)
thresh = cv2.bitwise_not(thresh)
plt.imshow(thresh, cmap="gray", vmin=0, vmax=255)
```

```python
kernel = np.ones((15,15), np.uint8)
img_dilation = cv2.dilate(thresh, kernel, iterations=1)
img_erode = cv2.erode(img_dilation,kernel, iterations=1)
img_erode = cv2.medianBlur(img_erode, 7)
plt.imshow(img_erode, cmap="gray", vmin=0, vmax=255)
```

```python
ret, labels = cv2.connectedComponents(img_erode)
label_hue = np.uint8(179 * labels / np.max(labels))
blank_ch = 255 * np.ones_like(label_hue)
labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)
labeled_img[label_hue == 0] = 0

plt.imshow(labeled_img)
print('The number of islands in the image is:', ret-1)
plt.show()
```

```python
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(im)
f.add_subplot(1,2, 2)
plt.imshow(labeled_img)
print('The number of islands in the image is:', ret-1)
```
