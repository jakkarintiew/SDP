# QC slide image against block image

### Import the necessary libraries

```python
from __future__ import print_function
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import measure
from skimage.morphology import skeletonize
from skimage import data
from skimage.util import invert
from skimage.io import imread

import sys
```

### Read processed images

```python
# Read all images
import os

slides = []
blocks = []

path = './img/Slides_processed/'
files = os.listdir(path)
for index, file in enumerate(files):
    # convert image from BGR to RGB
    img_sld = cv2.cvtColor(cv2.imread(os.path.join(path, file)), cv2.COLOR_BGR2RGB)
    slides.append(img_sld)

path = './img/Blocks_processed/'
files = os.listdir(path)
for index, file in enumerate(files):
    # convert image from BGR to RGB
    img_blk = cv2.cvtColor(cv2.imread(os.path.join(path, file)), cv2.COLOR_BGR2RGB)
    blocks.append(img_blk)
```

```python
# Select the sample pair that we are going to work with
sample_index = 1
img_slide = slides[sample_index].copy()
img_block = blocks[sample_index].copy()


# Display the images
plt.figure(figsize=(15,8))

plt.subplot(121)
plt.title('block preprocessed')
plt.imshow(img_block)

plt.subplot(122)
plt.title('slide preprocessed')
plt.imshow(img_slide)

plt.show()
```

### QC function
* The functions take in the preprocessed block-slide image pair, and return a series of scores that represent the "quality" of the slide image against the block image.
* Possible scoring methods:
    1. MSE
    2. Structure Similarity (SSIM)
    3. Boundaries
    4. Section completeness
    5. etc.

```python
def mse(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimension
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])

    # return the MSE, the lower the error, the more "similar"
    # the two images are
    return err

def qc(img_block, img_slide):
    img_block = cv2.cvtColor(img_block,cv2.COLOR_RGB2GRAY)
    img_slide = cv2.cvtColor(img_slide,cv2.COLOR_RGB2GRAY)
    # Add your code here
    
    # compute the mean squared error and structural similarity
    # index for the images
    score_mse = mse(img_block, img_slide)
    score_ssim = measure.compare_ssim(img_block, img_slide)

    return score_mse, score_ssim
```

```python
score_mse, score_ssim = qc(img_block, img_slide)

print("MSE: %.2f, SSIM: %.2f" % (score_mse, score_ssim))
```

```python
#matchTemplate methods
#For SQDIFF and SQDIFF_NORMED, the best matches are lower values. For all the other methods, the higher the better

#correlation coefficient
match = cv2.matchTemplate(img_slide, img_block, cv2.TM_CCOEFF)
print(match)

##correlation coefficient, normalised
match = cv2.matchTemplate(img_slide, img_block, cv2.TM_CCOEFF_NORMED)
print(match)

match = cv2.matchTemplate(img_slide, img_block, cv2.TM_SQDIFF)
print(match)

match = cv2.matchTemplate(img_slide, img_block, cv2.TM_SQDIFF_NORMED)
print(match)

match = cv2.matchTemplate(img_slide, img_block, cv2.TM_CCORR)
print(match)

match = cv2.matchTemplate(img_slide, img_block, cv2.TM_CCORR_NORMED)
print(match)
```

```python
#calcHist for both images, counts of data set into bins, looks at colour intensities
img_slide_hist=cv2.calcHist([img_slide],[0],None,[255],[0,255])
img_block_hist=cv2.calcHist([img_block],[0],None,[255],[0,255])

#compareHist methods which consists of (First compared histogram,Second compared histogram of the same size as first histogram, method)
#Correlation between the 2 histograms, ranges from [-1,1], 1 for perfect match
hist = cv2.compareHist(img_slide_hist, img_block_hist,cv2.HISTCMP_CORREL)
print(hist)

#Chi-square distance between histograms, [0, unbounded], 0 for perfect match
hist = cv2.compareHist(img_slide_hist, img_block_hist,cv2.HISTCMP_CHISQR)
print(hist)

#Intersection between the histograms, [0,unbounded], 0 for no match at all
hist = cv2.compareHist(img_slide_hist, img_block_hist,cv2.HISTCMP_INTERSECT)
print(hist)

#Bhattacharyya distance,[0,1], 0 for perfect match
hist = cv2.compareHist(img_slide_hist, img_block_hist,cv2.HISTCMP_BHATTACHARYYA)
print(hist)
```
